# Structural pauperism and the almshouses

We are trying to sort out a model in which several causes are building up in a particular way. These causes are:

1. We have data on the foreign/native composition of the poor in the almshouses in each county in New York State
2. We know flows - the number of people who enter and leave the almshouse during a year
3. Questions that concern us:
   A. Do homeless people tend to leave homelessess when care is better or when it is worse?
   B. How easily do people substitute between homelessness and work?
   C. How easily do people move across locations?
   These questions are interrelated, so maybe our model should suggest this.
4. Lots of people left New York during this time period, and lots came in from other states, while at the same time an enormous number came from overseas. Can we leverage entry and exit from our problem as ways of normalizing? 

### Basic utility

##### Agent type

To build a structural model, we will start by imagining that there is a type-dependent utility to being homeless and to working in a particular area. We will subscript each type by $a$, where $a$ can be some combination of male/female and foreign/native. That's all we have in the data, so we will go with that. 

##### Location

Each location is indexed by a county $c$, generically, where someone starts from influences utility, so we need to carry around a $c$ subscript on all of our quantities of interest. Locations offer utility to workers $w_c$, and benefits to being homeless which are $h_c$. These benefits can depend upon agent type, so we might write $w_{ac}$ and $b_{ac}$ instead. Ultimately, we should extend this even further to include time, so the worker benefits and the homeless benefits are $w_{act}$, and $b_{act}$, but we are going to try and start with a simple model where we don't think about such things right off the bat.  

#### Moving costs

These costs are generally given by $m_{c_1c_2}$ - the costs of moving from one county to another, which we will suppose are symmetric for now so we have something like a distance matrix with 0 on the diagonal. But they might also depend upon whether or not you are homeless, and also your current type. Hence, we should write $m^{as}_{c_1c_2}$, where $s=w,h$ are possible states. 

### Inclusive utilities

There are various ways to nest things, and one way is to nest within a location. We might do this if we expect that someone chooses a location first, and then chooses whether to work or be homeless there. How might we describe this situation? Given all of our above notations, we have (still leaving out time subscripts):

$$
U_{ac_ic_j}=\lambda \ln \left[\exp\left(\frac{w_{ac_j}-m^{as}_{c_ic_j}}{\lambda}\right)+\exp\left(\frac{h_{ac_j}-m^{as}_{c_ic_j}}{\lambda}\right)\right]
$$

This describes the utility that an agent who is currently in location $c_i$ would get if they were to move to $c_j$. Mind you, the agent does not yet know if he or she will be homeless or working in location $c_j$, so the agent gets the above expected utility from moving to $c_j$ at some cost. When there, to complete the story, the agent draws a frechet random variable upon arrival, and then decides, based on this idiosyncratic payoff component, to be homeless or to work. Note that this inclusive utility simplifies to:

$$
U_{ac_ic_j}=\lambda \ln \left[\exp\left(\frac{w_{ac_j}}{\lambda}\right)+\exp\left(\frac{h_{ac_j}}{\lambda}\right)\right]-m^{as}_{c_ic_j}
$$



The great thing about inclusive utilities is they give probabilities through something like Roy's Identity. So, we can get the probability that someone moving from $i$ to $j$ works in location $j$ is the derivative of the above utility with respect to $w_{ac_j}$:

$$
\frac{\partial U_{ac_ic_j}}{\partial w_{ac_j}}=\frac{\exp\left(\frac{w_{ac_j}}{\lambda}\right)}{\exp\left(\frac{w_{ac_j}}{\lambda}\right)+\exp\left(\frac{h_{ac_j}}{\lambda}\right)}=P_{w,acj}
$$
Note that because of the way in which moving costs linearize out of the utility they do not influence the decision to work or not in the new location. Also, we note that the decision to work or not doesn't depend upon the origin. 

Each agent at the beginning of every period chooses where to go to, and we can write the utility of a type $a$ agent in location $c_i$ as:

$$
\mathbf{U}_{ac_i} = \mu\ln\left[\sum_{c_j=1}^C \exp\left(\frac{U_{ac_ic_j}}{\mu}\right)\right]
$$

Where $U_{ac_ic_j}$ is as above is expected utility. 

### Entry and Exit

Now, we suppose that agents can enter or exit the game. Let's not worry too much about how all this goes for now, but suppose that there is no real nesting here, and agents get utility $e^{X_{c_i}}$ from leaving if they are currently in location $c_i$. Inclusive utility is then given by

$$
\mathcal{U}_{ac_i}=\ln \left[\exp(X_{c_i})+\exp(\mathbf{U_{ac_i}})\right]
$$

This expression can be used to derive the probability that an agent stays in state. For example, we find that a fraction of the inhabitants $P_{xc_i}$

$$
P_{xc_i} = \frac{e^{X_{c_i}}}{e^{X_{c_i}}+e^{\mathbf{U}_{ac_i}}}
$$

leave.

## Population outflows

Consider the derivative of $\mathbf{U}_{ac_i}$ with respect to $U_{ac_ic_k}$

$$
\frac{\partial \mathbf{U}_{ac_i}}{\partial U_{ac_ic_k}}=\frac{\exp\left(\frac{U_{ac_ic_k}}{\mu}\right)}{\sum_{c_j=1}^C \exp\left(\frac{U_{ac_ic_j}}{\mu}\right)}=P_{ac_ic_j}
$$

That is, this gives the probability that a person moves from $i$ to $j$. if we have starting populations $N_{ac_i}$, we can write the population in $j$
